# Geospatial Functions

```python
import mlflow
config = mlflow.models.ModelConfig(development_config='config.yaml')
dbutils.widgets.text('catalog', config.get('catalog'))
dbutils.widgets.text('schema', config.get('schema'))
```

```python
%sql
CREATE OR REPLACE FUNCTION `${catalog}`.`${schema}`.stations_within_distance(lat DOUBLE, long DOUBLE, dist DOUBLE)
RETURNS TABLE (
  code STRING,
  name STRING,
  latitude FLOAT,
  longitude FLOAT,
  is_public INT
)
RETURN
SELECT *
FROM `${catalog}`.`${schema}`.stations
WHERE st_dwithin(
  st_transform(st_geomfromtext(CONCAT('POINT(', longitude, ' ', latitude, ')'), 4326), 32188),
  st_transform(st_geomfromtext(CONCAT('POINT(', long, ' ', lat, ')'), 4326), 32188),
  dist
)
LIMIT 10;
```

```python
%sql
SELECT * FROM `${catalog}`.`${schema}`.stations_within_distance(45.5, -73.5, 10000) LIMIT 10
```
